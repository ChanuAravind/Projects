### Bagging : Random Forest

```python
params = {
    "n_estimators": [10,25,50,100,150,200],
    "max_depth" : [3, 5, 7],
    "max_leaf_nodes" : [15, 20, 25]
}

rfc = RFC(n_jobs = -1)
clf = RandomizedSearchCV(rfc, params, scoring = "accuracy", cv=3, n_jobs = -1, verbose = 1)

clf.fit(x_sm, y_sm)
```

```python
res = clf.cv_results_

for i in range(len(res["params"])):
  print(f"Parameters:{res['params'][i]} Mean_score: {res['mean_test_score'][i]} Rank: {res['rank_test_score'][i]}")
```

```python
print(clf.best_estimator_)
```

```python
rf = clf.best_estimator_

rf.fit(x_sm, y_sm)

print("Model acc",rf.score(X_test, y_test))
```

### Boosting : GBDT

```python

gbc = GBC(n_estimators=150, learning_rate=0.2, max_depth=4, random_state=0, verbose = 1).fit(x_sm, y_sm)

print(f"Training accuracy:{gbc.score(x_sm, y_sm)}\nTest Accuracy: {gbc.score(X_test, y_test)}")
```
